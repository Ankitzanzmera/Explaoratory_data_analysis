```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./vaccination-data.csv')
df.head()
```

```python
temp_data = df[['TOTAL_VACCINATIONS','COUNTRY']].nlargest(n = 10,columns='TOTAL_VACCINATIONS')
plt.figure(figsize = [15,5])
sns.barplot(data = temp_data,x = 'COUNTRY',y= 'TOTAL_VACCINATIONS')
plt.title('TOP 10 Countries with Total number of Vaccinations')
for i in range(len(temp_data.TOTAL_VACCINATIONS)):
    plt.text(i,temp_data.TOTAL_VACCINATIONS.iloc[i],str(temp_data.TOTAL_VACCINATIONS.iloc[i]),ha = 'center')
plt.show()
```

```python
temp_data = df[['TOTAL_VACCINATIONS_PER100','COUNTRY']].nlargest(n = 10,columns='TOTAL_VACCINATIONS_PER100')
plt.figure(figsize = [15,5])
sns.barplot(data = temp_data,x = 'COUNTRY',y= 'TOTAL_VACCINATIONS_PER100')
plt.title('No of Doses per 100 Person')
for i in range(len(temp_data.TOTAL_VACCINATIONS_PER100)):
    plt.text(i,temp_data.TOTAL_VACCINATIONS_PER100.iloc[i],str(temp_data.TOTAL_VACCINATIONS_PER100.iloc[i]),ha = 'center')
plt.show()
```

```python
plt.pie(temp_data['TOTAL_VACCINATIONS'],labels=temp_data['COUNTRY'],autopct = '%.1f%%')
plt.title('Top 10 Countries With Percentage of Total_vaccinations')
plt.show()
```

```python
data = df[['COUNTRY','PERSONS_VACCINATED_1PLUS_DOSE']].nlargest(10,columns = 'PERSONS_VACCINATED_1PLUS_DOSE')
plt.figure(figsize = [15,5])
sns.barplot(data = data,x = 'COUNTRY',y = 'PERSONS_VACCINATED_1PLUS_DOSE')
plt.title(' Country Wise Number of Persons Who Takes atleast one Dose')
for i in range(len(data.PERSONS_VACCINATED_1PLUS_DOSE)):
    plt.text(i,data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i],str(data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i]),ha = 'center')
plt.show()
```

```python
data = df[['COUNTRY','PERSONS_VACCINATED_1PLUS_DOSE_PER100']].nlargest(10,columns = 'PERSONS_VACCINATED_1PLUS_DOSE_PER100')
plt.figure(figsize = [17,5])
sns.barplot(data = data,x = 'COUNTRY',y = 'PERSONS_VACCINATED_1PLUS_DOSE_PER100')
plt.title('number of people who have received at least one dose of the vaccine per 100 people in the country')
for i in range(len(data.PERSONS_VACCINATED_1PLUS_DOSE_PER100)):
    plt.text(i,data.PERSONS_VACCINATED_1PLUS_DOSE_PER100.iloc[i],str(data.PERSONS_VACCINATED_1PLUS_DOSE_PER100.iloc[i]),ha = 'center')
plt.show()
```

```python
data = df[['COUNTRY','TOTAL_VACCINATIONS','PERSONS_VACCINATED_1PLUS_DOSE']].nlargest(10,columns = 'PERSONS_VACCINATED_1PLUS_DOSE')
plt.figure(figsize = [20,8])
plt.bar(data['COUNTRY'],data['TOTAL_VACCINATIONS'],label='TOTAL_VACCINATIONS')
plt.bar(data['COUNTRY'],data['PERSONS_VACCINATED_1PLUS_DOSE'],label = 'PERSONS_VACCINATED_1PLUS_DOSE')
plt.title('Country Wise Tota Vaccination with Respect to person who take alleast one Dose of Vaccination')
for i in range(len(data.COUNTRY)):
    plt.text(i,data.TOTAL_VACCINATIONS.iloc[i],str(data.TOTAL_VACCINATIONS.iloc[i]),ha = 'center')
    plt.text(i,data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i],str(data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i]),ha = 'center',color = 'white')
plt.legend()
plt.show()
```

```python
data = df[['COUNTRY','PERSONS_FULLY_VACCINATED']].nlargest(10,columns = 'PERSONS_FULLY_VACCINATED')
plt.figure(figsize = [15,5])
sns.barplot(data = data,x = 'COUNTRY',y = 'PERSONS_FULLY_VACCINATED')
plt.title('Top Country with most Number of Person Who are Fully Vaccinated')
for i in range(len(data.PERSONS_FULLY_VACCINATED)):
    plt.text(i,data.PERSONS_FULLY_VACCINATED.iloc[i],str(data.PERSONS_FULLY_VACCINATED.iloc[i]),ha = 'center')
plt.show()
plt.show()
```

```python
data = df[['COUNTRY','PERSONS_VACCINATED_1PLUS_DOSE_PER100']].nlargest(10,columns = 'PERSONS_VACCINATED_1PLUS_DOSE_PER100')
plt.figure(figsize = [15,5])
sns.barplot(data = data,x = 'COUNTRY',y = 'PERSONS_VACCINATED_1PLUS_DOSE_PER100')
plt.title('number of people who have received all required doses of the vaccine per 100 people in the country')
for i in range(len(data.PERSONS_VACCINATED_1PLUS_DOSE_PER100)):
    plt.text(i,data.PERSONS_VACCINATED_1PLUS_DOSE_PER100.iloc[i],str(data.PERSONS_VACCINATED_1PLUS_DOSE_PER100.iloc[i]),ha = 'center')
plt.show()
plt.show()
```

```python
data = df[['COUNTRY','TOTAL_VACCINATIONS','PERSONS_VACCINATED_1PLUS_DOSE','PERSONS_FULLY_VACCINATED']].nlargest(10,columns = 'PERSONS_VACCINATED_1PLUS_DOSE')
plt.figure(figsize = [20,8])
plt.bar(data['COUNTRY'],data['TOTAL_VACCINATIONS'],label = 'Total Vaccination')
plt.bar(data['COUNTRY'],data['PERSONS_VACCINATED_1PLUS_DOSE'],label = 'Person Who Takes Atleast One Dose',color = 'b',bottom = 0)
plt.bar(data['COUNTRY'],data['PERSONS_FULLY_VACCINATED'],label = 'Fully Vaccinated',color = 'r',bottom = data['PERSONS_VACCINATED_1PLUS_DOSE'])

for i in range(len(data.COUNTRY)):
    plt.text(i,data.TOTAL_VACCINATIONS.iloc[i]+10000009,str(data.TOTAL_VACCINATIONS.iloc[i]),ha = 'center')
    plt.text(i,data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i],str(data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i]),ha = 'center',color = 'white')
    plt.text(i,data.PERSONS_FULLY_VACCINATED.iloc[i] + data.PERSONS_VACCINATED_1PLUS_DOSE.iloc[i],str(data.PERSONS_FULLY_VACCINATED.iloc[i]),ha = 'center',color = 'white')

plt.title('Country Wise Tota Vaccination with Respect to person who take atleast one Dose of Vaccination,and No of Total Vaccinated Person')
plt.legend()
plt.show()
```

```python
data = df.groupby(by = ['NUMBER_VACCINES_TYPES_USED']).count().reset_index()
plt.pie(x = data['COUNTRY'],labels = data['NUMBER_VACCINES_TYPES_USED'],autopct = '%.1f%%',wedgeprops = {'edgecolor':'black'})
plt.title('% of Countries By Number of vaccine Type They used')
plt.show()
```

```python
df['PERSON_NOT_VACCINATES_ATALL'] = df['TOTAL_VACCINATIONS'] - df['PERSONS_VACCINATED_1PLUS_DOSE']
```

```python
data = df[['COUNTRY','PERSON_NOT_VACCINATES_ATALL']].nlargest(10,columns = 'PERSON_NOT_VACCINATES_ATALL')
plt.figure(figsize = [17,5])
sns.barplot(data = data,x = 'COUNTRY',y = 'PERSON_NOT_VACCINATES_ATALL')
plt.title("Country WIse Most Number no of person who didn't take atleast one dose")
for i in range(len(data.PERSON_NOT_VACCINATES_ATALL)):
    plt.text(i,data.PERSON_NOT_VACCINATES_ATALL.iloc[i],str(data.PERSON_NOT_VACCINATES_ATALL.iloc[i]),ha = 'center')
plt.show()
```

```python
df['TOOK_DOSE1_BUT_NOT_DOSE2'] = df['PERSONS_VACCINATED_1PLUS_DOSE'] - df['PERSONS_FULLY_VACCINATED']
```

```python
data = df[['COUNTRY','TOOK_DOSE1_BUT_NOT_DOSE2']].nlargest(10,columns = 'TOOK_DOSE1_BUT_NOT_DOSE2')
plt.figure(figsize = [17,5])
sns.barplot(data = data,x = 'COUNTRY',y = 'TOOK_DOSE1_BUT_NOT_DOSE2')
plt.title("Country WIse Most Number no of person who Took First Dose But Not Second Dose")
for i in range(len(data.TOOK_DOSE1_BUT_NOT_DOSE2)):
    plt.text(i,data.TOOK_DOSE1_BUT_NOT_DOSE2.iloc[i],str(data.TOOK_DOSE1_BUT_NOT_DOSE2.iloc[i]),ha = 'center')
plt.show()
```

```python
data = df[['COUNTRY','PERSONS_FULLY_VACCINATED','PERSONS_BOOSTER_ADD_DOSE']].nlargest(10,columns = ['PERSONS_FULLY_VACCINATED','PERSONS_BOOSTER_ADD_DOSE'])
plt.figure(figsize = [17,5])
plt.bar(data['COUNTRY'],data['PERSONS_FULLY_VACCINATED'],label = 'PERSONS_FULLY_VACCINATED')
plt.bar(data['COUNTRY'],data['PERSONS_BOOSTER_ADD_DOSE'],label = 'WHO Took Booster/Additional Dose')
for i in range(len(data.COUNTRY)):
    plt.text(i,data.PERSONS_FULLY_VACCINATED.iloc[i]+19000000,str(data.PERSONS_FULLY_VACCINATED.iloc[i]),ha = 'center')
    plt.text(i,data.PERSONS_BOOSTER_ADD_DOSE.iloc[i],str(data.PERSONS_BOOSTER_ADD_DOSE.iloc[i]),ha = 'center',color = 'white')
plt.legend()
plt.show()
```
